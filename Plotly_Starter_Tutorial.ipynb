## **Plotly**

Advantage : Excellent interactive visualizations, in the form of HTML files, where many wide range of features are available, like zooming in etc.

Disadvantage : Can't work on real time dynamic, live data. Only works on static data source.

This disdvantage is eradicated by Dash

```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
import plotly.figure_factory as ff
from plotly import tools
```

```python
match = pd.read_csv("C:\\Users\\BIPLAB\\ML_WinterTraining\\matches.csv")
delivery = pd.read_csv("C:\\Users\\BIPLAB\\ML_WinterTraining\\deliveries.csv")

ipl = delivery.merge(match,left_on='match_id',right_on='id')
ipl.head()
```

### Scatter plot b/w Batsman Avg(X axis) and Batsman Strike Rate(Y axis) of top 50 batsmen

```python
top50 = ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(50).index.tolist()
new_ipl = ipl[ipl['batsman'].isin(top50)]
new_ipl
```

```python
# Strike Rate
runs = new_ipl.groupby('batsman')['batsman_runs'].sum()
balls = new_ipl.groupby('batsman')['batsman_runs'].count()

sr = (runs/balls)*100
sr= sr.reset_index()
sr
```

```python
# Avg
out = ipl[ipl.player_dismissed.isin(top50)]
nouts = out.player_dismissed.value_counts()

avg = runs/nouts
avg = avg.reset_index()
avg.rename(columns={'index':'batsman',0:'avg'},inplace=True)
avg=avg.merge(sr,on='batsman')
avg
```

```python
# For 1 graph, 1 trace. For n graphs, n traces
trace = go.Scatter(x=avg.avg,y=avg.batsman_runs,mode='markers',text=avg.batsman,marker={'color':'#00a65a','size':12})
data = [trace]  # Why list ? Because we might want to plot multiple graphs in same plot
layout = go.Layout(title='BATSMAN AVG VS STRIKE RATE',xaxis={'title':'Batsman Avg'},yaxis={'title':'Batsman Strike Rate'})

fig = go.Figure(data=data,layout=layout)
pyo.plot(fig,filename='plotly_scatter.html')
```

### There are 3 types of Bar Graphs in Plotly :

#### 1. Nested Graph
#### 2. Stacked Graph
#### 3. Overlayed Graph

### Suppose we want to show the total runs scored per innings per batsman

```python
iw = new_ipl2.groupby(['batsman','inning'])['batsman_runs'].sum().reset_index()
# Doing for 2 innings
mask1 = iw['inning']==1
mask2 = iw['inning']==2
one = iw[mask1]
two = iw[mask2]

one.rename(columns={'batsman_runs':'1st Inning'},inplace=True)
two.rename(columns={'batsman_runs':'2nd Inning'},inplace=True)
final = one.merge(two,on='batsman')[['batsman','1st Inning','2nd Inning']]

final
```

```python
# Multiple bar graphs(the types)
trace1 = go.Bar(x=final.batsman,y=final['1st Inning'],marker={'color':'#00a65a'},name='1st Innings')
trace2 = go.Bar(x=final.batsman,y=final['2nd Inning'],marker={'color':'#a6a65a'},name='2nd Innings')
data = [trace1,trace2]
layout = go.Layout(title='INNING WISE PERFORMANCE',xaxis={'title':'Season'},yaxis={'title':'Performance'},barmode='stack')

# barmode : default - nested/side by side stacked
# barmode : overlay - stacked vertically such that the one with higher amount covers the one with smaller.
# barmode : stack - stacked vertically, no covering up
 
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig,filename='plotly_line_types.html')
```

### Distplot of average runs vs strike rate

### Histogram of strike rates of batsmen who have played more than 150 balls

```python
x = delivery.groupby('batsman')['batsman_runs'].count()>150
x = x[x].index.tolist()

new = delivery[delivery.batsman.isin(x)]

runs = new.groupby('batsman')['batsman_runs'].sum()
balls = new.groupby('batsman')['batsman_runs'].count()

sr = (runs/balls)*100
sr = sr.reset_index()

sr
```

```python
trace = go.Histogram(x=sr.batsman_runs,xbins={'size':2})
data=[trace]
layout = go.Layout(title='STRIKE RATE ANALYSIS',xaxis={'title':'Strike Rate'})

fig = go.Figure(data=data,layout=layout)
pyo.plot(fig,filename="plotly_histogram.html")
```

### Heatmaps

```python
six = delivery[delivery.batsman_runs==6]
six = six.groupby(['batting_team','over'])['batsman_runs'].count().reset_index()
six
```

```python
trace = go.Heatmap(x=six.batting_team,y=six.over,z=six.batsman_runs)
data=[trace]
layout = go.Layout(title='SIX HEATMAP')

fig = go.Figure(data=data,layout=layout)
pyo.plot(fig,filename="plotly_heatmaps.html")
```

```python
# Plotting two heatmaps side by side : SIXES and DOTS

dots = delivery[delivery.batsman_runs==0]
dots = dots.groupby(['batting_team','over'])['batsman_runs'].count().reset_index()
dots
```

```python
trace1 = go.Heatmap(x=six.batting_team,y=six.over,z=six.batsman_runs.values.tolist())
trace2 = go.Heatmap(x=dots.batting_team,y=dots.over,z=dots.batsman_runs.values.tolist())
data=[trace1,trace2]
layout = go.Layout(title='SIX HEATMAP')

fig = tools.make_subplots(rows=1,cols=2,subplot_titles=["6's","0's"],shared_yaxes=True)   
# rows = no. of rows we want, cols = no.of columns we want, shared_yaxes = True : same y axis to be used, different x axis

fig.append_trace(trace1,1,1)  # 1st row,1st column
fig.append_trace(trace2,1,2)  # 1st row,2nd column
pyo.plot(fig,filename="plotly_heatmaps.html")
```
